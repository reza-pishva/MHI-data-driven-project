```python
import mysql.connector
import pandas as pd

# اتصال به دیتابیس
try:
    connection = mysql.connector.connect(
        host='127.0.0.1',
        port=3306,
        user='root',
        password='',
        database='dsas'
    )
    print("اتصال به دیتابیس با موفقیت انجام شد.")
except mysql.connector.Error as err:
    print(f"خطا در اتصال به دیتابیس: {err}")
    exit()

# کوئری SELECT برای 1000 رکورد اول
query = """
SELECT 
    `id`, `AssetID_8341`, `AssetID_8342`, `AssetID_8343`, `AssetID_8344`, 
    `AssetID_8346`, `AssetID_9286`, `AssetID_9287`, `unitID`, `model_name`, 
    `model_result`, `model_value`, `DateTime`, `RecordDate`, `RecordTime`, 
    `TimeStamps`, `created_at`, `updated_at`
FROM `results_table_mhi_lube_oil_g11`
LIMIT 1000
"""

# خواندن داده‌ها با pandas
try:
    df = pd.read_sql(query, connection)
    print(f"{len(df)} رکورد با موفقیت خوانده شد.")
except Exception as e:
    print(f"خطا در اجرای کوئری: {e}")
    connection.close()
    exit()

# بستن اتصال
connection.close()

# ذخیره در فایل اکسل
output_file = 'dbscan_model1.xlsx'
try:
    df.to_excel(output_file, index=False)
    print(f"فایل اکسل با موفقیت در مسیر زیر ذخیره شد: {output_file}")
except Exception as e:
    print(f"خطا در ذخیره فایل اکسل: {e}")
```
